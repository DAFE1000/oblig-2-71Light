```python
import matplotlib.pyplot as plt
import numpy as np

#f(x) = e^(-x/4) * tan^-1(x)
def f1(x):
    return np.e ** (-x/4) * np.arctan(x)

#f'(x), den deriverte av f(x)
def fd1(x):
    return np.exp(-x/4) * (1/(x**2 + 1) - (1/4)*np.arctan(x))

#f''(x), den deriverte av den deriverte av f(x)
def fdd1(x):
    return np.exp(-x/4) * (-2*x/(x**2 + 1)**2) - (1/2)*(1/(x**2 + 1)) + (1/16)*np.arctan(x)

#ligningen tan^-1(x) - 4/(x**2+1), bare at den er gjort til en funksjon istedet for et stykke slik at den er lettere og plotte og for å vise at når den krysser f'(x) så er y=0 som vil si at de har samme verdi. Siden f'(x) viser stingigen til f(x) vil det si at når y=0 så er enten vil på ett toppunkt eller bunnpunkt.
def f2(x):
    return np.arctan(x) - 4/(x**2+1)

x = np.linspace(-1, 5, 100)

y1 = f1(x)
yd1 = fd1(x)
y2 = f2(x)

x0 = 1

while("true"):
    x1 = x0 - fd1(x0)/fdd1(x0)
    if (round(fd1(x1), 5) == 0):
        print("funksjon har høyest verdi på x =", round(x1, 4), "rundet ned til 4 desimaler")
        break
    else:
        x0 = x1

plt.scatter(x0, f1(x0), label="MAX", color="blue")

plt.plot(x, y1, label="f(x)", color="green")
plt.plot(x, yd1, label ="fd(x)", color="limegreen")
plt.plot(x, y2, label="g(x)", color="red")

plt.xlabel("x")
plt.ylabel("y")
plt.grid("true")
plt.show
```
